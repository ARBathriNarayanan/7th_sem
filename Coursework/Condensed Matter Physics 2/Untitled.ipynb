```python
import numpy as np
import scipy as sp
from scipy.constants import k
import matplotlib.pyplot as plt

T=np.linspace(4,5000,5000)
delta1=0.01
delta2=0.05
delta3=0.1
kev=k/(sp.constants.e)
sigma0=5
sigma1=sigma0*np.exp(-delta1/(kev*T))
sigma2=sigma0*np.exp(-delta2/(kev*T))
sigma3=sigma0*np.exp(-delta3/(kev*T))

#plt.plot(T,sigma1,color="black",label="D=0.01")
#plt.plot(T,sigma2,label="D=0.05")
plt.plot(T,sigma3,label="D=0.1")
plt.legend()
plt.xlabel('T')
plt.ylabel('Sigma')
plt.title("Plot of T vs Sigma")
plt.show()

def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
```

```python
T=np.linspace(4,2000,5000)
ds1dt=sigma0*np.exp(-delta1/(kev*T))*(delta1/(kev*(T**2)))
ds2dt=sigma0*np.exp(-delta2/(kev*T))*(delta2/(kev*(T**2)))
ds3dt=sigma0*np.exp(-delta3/(kev*T))*(delta3/(kev*(T**2)))
#plt.plot(T,ds1dt,color="black",label="D=0.01")
#plt.plot(T,ds2dt,label="D=0.05")
plt.plot(T,ds3dt,label="D=0.1")
plt.legend()
plt.xlabel('T')
plt.ylabel('Sigma')
plt.title("Plot of T vs d(Sigma)/dt")
plt.show()
```

```python
T=np.linspace(4,1000,5000)
d2s1dt2=sigma0*np.exp(-delta1/(kev*T))*((delta1**2)/((kev**2)*(T**4)))-2*sigma0*np.exp(-delta1/(kev*T))*(delta1/(kev*(T**3)))
d2s2dt2=sigma0*np.exp(-delta2/(kev*T))*((delta2**2)/((kev**2)*(T**4)))-2*sigma0*np.exp(-delta2/(kev*T))*(delta2/(kev*(T**3)))
d2s3dt2=sigma0*np.exp(-delta3/(kev*T))*((delta3**2)/((kev**2)*(T**4)))-2*sigma0*np.exp(-delta3/(kev*T))*(delta3/(kev*(T**3)))
#plt.plot(T,d2s1dt2,color="red",label="D=0.01")
#plt.plot(T,d2s2dt2,label="D=0.05")
plt.plot(T,d2s3dt2,label="D=0.1")
plt.legend()
plt.xlabel('T')
plt.ylabel('Sigma')
plt.title("Plot of T vs d^2(Sigma)/dt^2")
plt.show()
```
