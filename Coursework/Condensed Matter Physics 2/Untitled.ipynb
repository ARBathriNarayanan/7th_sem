```python
import numpy as np
import scipy as sp
from scipy.constants import k
import matplotlib.pyplot as plt

T=np.linspace(4,5000,5000)
delta2=0.1
delta1=0.05
delta3=0.5
kev=k/(sp.constants.e)
sigma0=5
sigma1=sigma0*np.exp(-delta1/(kev*T))
sigma2=sigma0*np.exp(-delta2/(kev*T))
sigma3=sigma0*np.exp(-delta3/(kev*T))

plt.plot(T,sigma1,color="black",label="D=0.05")
plt.plot(T,sigma2,label="D=0.1")
plt.plot(T,sigma3,label="D=0.5")
plt.legend()
plt.xlabel('T')
plt.ylabel('Sigma')
plt.title("Plot of T vs Sigma")
plt.show()
```
