```python
import numpy as np
from matplotlib import pyplot as plt
gold_winners = [9.95,10.14,10.06,10.25,9.99,9.92,9.96,9.84,9.87,9.85,9.69,9.63,9.81,9.80,9.784]
silver_winners=[10.04,10.24,10.08,10.25,10.19,9.97,10.02,9.89,9.99,9.86,9.89,9.75,9.89,9.84,9.789]
x=np.linspace(1968,2024,15)
print(x)
plt.plot(x,gold_winners,label="Time of gold medal winner")
plt.plot(x,silver_winners,label="Time of silver medal winner")
plt.xlabel("Year")
plt.ylabel("Time in seconds")
plt.title("Time set by gold and silver medallists in Olympics")
plt.legend()
plt.savefig("time.png")
```

```python
delta=[]
for i in range(len(x)):
    delta.append(-gold_winners[i]+silver_winners[i])
plt.plot(x,delta)
```
