```python
import csv 

def collector (filename):
    with open(filename) as file:
        reader = csv.reader(file, delimiter=' ')
        for row in reader:
            raw.append(row)
    del raw[0]
    l,c=[],[]
    for i in range (len(raw)):
        if (raw[i][0]==""):
         l.append(raw[i][1])
         c.append(raw[i][3])
        else:
         l.append(raw[i][0])
         c.append(raw[i][2])
    return l,c
```

```python
from ROOT import TCanvas, TGraph, TLegend, gStyle
%jsroot on

canvas = TCanvas("canvas", "Wavelength vs Intensity", 800, 600)

# Create a TGraph
n_points = len(l4940025)
graph = TGraph(n_points)
for i in range(n_points):
    graph.SetPoint(i, l4940025[i], c4940025[i])

# Style the graph
graph.SetTitle("Wavelength vs Intensity;Wavelength;Intensity")  # Title and axis labels
graph.SetLineColor(4)  # Set line color (4 = blue)
graph.SetLineWidth(2)  # Set line width
graph.SetMarkerStyle(21)  # Set marker style
graph.SetMarkerSize(1)
graph.SetMarkerColor(2)  # Set marker color (2 = red)

# Draw the graph
graph.Draw("APL")  # A = axis, P = points, L = line

# Add a legend
legend = TLegend(0.1, 0.7, 0.48, 0.9)  # Legend position (x1, y1, x2, y2)
legend.AddEntry(graph, "Ring anode 0.49 Bar, 400V, 25mA", "lp") 
legend.Draw()

# Display the canvas
canvas.Draw()
```
